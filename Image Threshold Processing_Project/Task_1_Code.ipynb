```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load image
image = cv.imread('number.jpg')

if image is None:
    print("Error: Unable to load the image.")
else:
    # BGR to RGB conversion
    image = cv.cvtColor(image, cv.COLOR_BGR2RGB)

    # Convert image data type to float
    rgb_image = image / 255.0

    # Formula
    B = rgb_image[:, :, 2]  # Red channel in RGB
    R = rgb_image[:, :, 0]  # Blue channel in RGB
    G = rgb_image[:, :, 1]  # Green channel in RGB
    b = 3 * B / (R + G + B)

    # Threshold value
    threshold_value = 0.8
    binary_image = (b > threshold_value)
    inverted_image = 1 - binary_image

    # Original image
    plt.figure(figsize=(14, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(image)
    plt.title("Original Image")

    # Inverted image
    plt.subplot(1, 2, 2)
    plt.imshow(inverted_image, cmap='gray')
    plt.title("Thresholded Image")
    plt.show()
    
    #Binary Image
    plt.figure(figsize=(11.5, 5))
    plt.subplot(1, 2, 2)
    plt.imshow(binary_image, cmap='gray')
    plt.title("binary_image Image")
    plt.show()
```
